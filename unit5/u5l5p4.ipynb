```
# Exploring tweet word count
z = df_2014.groupby(by=['cntry'], as_index=False).count(
).sort_values(by='sclmeet', ascending=True)

plt.figure(figsize=(8, 5))
sns.set_style("whitegrid")
ax = sns.barplot(x=z['cntry'], y=z['sclmeet'], data=z)
plt.xlabel("\nCountry")
plt.ylabel("Count")
plt.title("Social Activity by Country\n")
plt.show()
```
